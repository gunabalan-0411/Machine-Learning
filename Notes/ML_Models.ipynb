# Supervised Learning

## SVM

Hyper parameter;

gamma : tells how much decision line is complex
      : Lower then, broad decision: underfit
      : higher then, narrow decision: overfit

C     : Regularization parameter
      : low value means low penalty and underfit chances
      : higher value, high penalty and overfit chances

Kernel: method to transform data to make it separable, Many datasets are not linearly separable in original space, So we use a kernel to allow non-linear boundaries.
      : RBF Radial Basis Function Also called Gaussian kernel
      : Intuition:If two points are close → kernel output is high (very similar)
      : If far → output near 0
      : So RBF allows SVM to draw boundaries like:
      : circles / curves / complex shapes
